```python
import data_methods as dm
import matplotlib.pyplot as plt
import numpy as np

from visualization_methods import VisualizationMethods as vm

dfs = dm.get_refacts_dataframes()
```

```python
fig, ax = plt.subplots(figsize=(16, 5))
fig.set_facecolor('white')

keys = list(dfs.keys())
dot = False

for metric in ['lcom', 'cbo', 'wmc', 'rfc']:
    values = [np.mean(dfs[k]['classes'][metric]) for k in keys]
    line_style = '--' if dot else '-'
    ax.plot(keys, values, ls=line_style, label=f'class - {metric.upper()}')
    dot = not dot

for metric in ['fanin', 'cbo', 'loc', 'rfc']:
    values = [np.mean(dfs[k]['methods'][metric]) for k in keys]
    line_style = '--' if dot else '-'
    ax.plot(keys, values, ls=line_style, label=f'metric - {metric.upper()}')
    dot = not dot

plt.legend()
plt.tight_layout()
# plt.savefig('metrics_progress.pdf')
plt.show()
```
